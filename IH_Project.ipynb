```python
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

##Reading json files

```python
import json
with open(r"m.json","r") as read_file:
  m = json.load(read_file)
```

##Printing json files

```python
m
```

##Specifying model type and model name




```python
model_type="bert"
model_name= "bert-base-cased"
if model_type == "bert":
    model_name = "bert-base-cased"

elif model_type == "roberta":
    model_name = "roberta-base"

elif model_type == "distilbert":
    model_name = "distilbert-base-cased"

elif model_type == "distilroberta":
    model_type = "roberta"
    model_name = "distilroberta-base"

elif model_type == "electra-base":
    model_type = "electra"
    model_name = "google/electra-base-discriminator"

elif model_type == "electra-small":
    model_type = "electra"
    model_name = "google/electra-small-discriminator"

elif model_type == "xlnet":
    model_name = "xlnet-base-cased"
```

##Configuring Question Answer model

```python
model_args = QuestionAnsweringArgs()
model_args.train_batch_size = 16
model_args.evaluate_during_training = False
model_args.n_best_size = 3
model_args.num_train_epoch = 5
```

```python
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "use_cached_eval_features": True,
    "output_dir": f"outputs/{model_type}",
    "best_model_dir": f"outputs/{model_type}/best_model",
    "evaluate_during_training": False,
    "max_seq_length": 128,
    "num_train_epochs": 10,
    "evaluate_during_training_steps": False,
    "wandb_project": "Question Answer Application",
    "wandb_kwargs": {"name": model_name},
    "save_model_every_epoch": False,
    "save_eval_checkpoints": False,
    "n_best_size":3,
    "use_cuda" : False,
    "no_deprecation_warning":True,
    # "use_early_stopping": True,
    # "early_stopping_metric": "mcc",
    # "n_gpu": 2,
    # "manual_seed": 4,
    # "use_multiprocessing": False,
    "train_batch_size": 128,
    "eval_batch_size": 64,
    # "config": {
    #     "output_hidden_states": True
    # }
}
```

##Initializing Question Answer model

```python
model = QuestionAnsweringModel(model_type,model_name,args = train_args)
```

##Removing output files

```python
!rm -rf outputs
```

##Training model

```python
model.train_model(m)
```

##Evaluating model

```python
results, texts = model.eval_model(m)
```

##Making predictions

```python
to_predict = [
    {
        "context": "Corrections on grade card /PDC",
        "qas": [
            {
                "question": "What if I need to know my correct CGPA?",
                "id": "1",
            }
        ],
    }
]
```

##Printing output

```python
answer,probabilities = model.predict(to_predict)
print(answer)
```
